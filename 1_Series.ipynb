```python
# Creating pandas series
import numpy as np
import pandas as pd

labels = ['a','b','c']
list = [1,2,3]
arr = np.array([3,4,5])
dict = {'d': 34, 'e': 35, 'f': 87}
```

```python
pd.Series(list, index = labels)
```

```python
pd.Series(arr, index = list)
```

```python
pd.Series(dict)
```
