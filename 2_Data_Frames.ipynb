```python
## From Dictionary


import numpy as np
import pandas as pd


data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}


df1 = pd.DataFrame(data)
df1
```

```python
## From list

data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]

col_header = ["Name", "Age", "City", "Salary"]

df2 = pd.DataFrame(data_list, columns=col_header)
df2
```
