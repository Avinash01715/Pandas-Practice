# Basic Operations on DataFrames

```python
#sample data
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})
df
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
## broadcasting on column
df[["A","B"]] + 10
```

```python
# applying functions

def square(x):
    return x**2

df["B"]=df["B"].apply(square)
df
```

```python
df["D"] = df["A"].apply( lambda x : x + 10)
df
```
