# Missing Data

```python
import numpy as np
import pandas as pd
```

## Finding 

```python
# Sample data dictionary
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

## Removing

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh= 3)
```

## Filling

```python
df
```

```python
df.fillna(1)
```

```python
v = {'A': 34, 'B':22, 'C': 99, 'D': 88}

df.fillna(value= v)
```

```python
df.fillna(df.mean())
```
