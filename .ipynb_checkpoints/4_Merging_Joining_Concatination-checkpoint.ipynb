```python
import numpy as np
import pandas as pd
```

## Merging two dataframe

```python
# Sample data fdataframe
df1 = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
df2 = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})

#Employee_id common
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='employee_id',how='inner')
```

```python
pd.merge(df1, df2, on='employee_id',how='outer')
```

```python
pd.merge(df1, df2, on='employee_id',how='left')
```

```python
pd.merge(df1, df2, on='employee_id',how='right')
```

# Concatination of two data frames

```python
# sample data frames
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], axis = 1)
```

# Joining two Data frames

```python
#Sample data frames
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='inner')
```
